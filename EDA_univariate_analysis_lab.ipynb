# Understanding Product Categories

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# file path
file_path = r'C:\Users\joann\OneDrive - University of Twente\Desktop\IRONHACK\CourseWork\Week5\Week1\Day1\Amazon_dataset\amz_uk_processed_data.csv'

# Loading the dataset
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
df.columns
```

```python
# Generate a frequency table for the product category
category_frequency = df['categoryName'].value_counts()
category_frequency
```

```python
# Print the top 5 most listed product categories
top_5_categories = category_frequency.head(5)
print("Top 5 most listed product categories:")
print(top_5_categories)
```

```python
# Display the distribution of products across different categories using a bar chart
top_categories = category_frequency.head(10)
plt.figure(figsize=(8, 8))
top_categories.plot(kind='barh', legend=None)  # Use 'barh' for horizontal bar plot
plt.title('Proportions of Top Product Categories')
plt.xlabel('Count')  # Set the x-axis label to 'Count'
plt.ylabel('Category')  # Set the y-axis label to 'Category'
plt.show()
```

# Delving into Product Pricing

```python
# Calculate the mean, median, and mode for the price of products
mean_price = round(df['price'].mean(), 2)
median_price = round(df['price'].median(), 2)
mode_price = round(df['price'].mode()[0], 2)

print("Mean price:", mean_price)
print("Median price:", median_price)
print("Mode price:", mode_price)
```

```python
# Determine the variance, standard deviation, range, and interquartile range for product price
variance_price = round(df['price'].var(), 2)
std_dev_price = round(df['price'].std(), 2)
price_range = df['price'].max() - df['price'].min()
interquartile_range_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print("\nVariance Price:", variance_price)
print("Standard Deviation Price:", std_dev_price)
print("Price Range:", price_range)
print("Interquartile Range Price:", interquartile_range_price)
```

```python
# Plot a histogram to visualize the distribution of product prices with adjusted bins
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=20, color='skyblue', edgecolor='black') 
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Use a box plot to showcase the spread and potential outliers in product pricing
plt.figure(figsize=(11, 6))
plt.boxplot(df['price'], vert=False)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.grid(True)
plt.show()
```

# Unpacking Product Ratings

```python
# Calculate the mean, median, and mode for the stars of products
mean_rating = round(df['stars'].mean(),2)
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print("Mean rating:", mean_rating)
print("Median rating:", median_rating)
print("Mode rating:", mode_rating)
```

```python
#How do customers generally rate products? Is there a common trend?
#Ans: Customers rate products very lowly, with a mean of about 2 stars
```

```python
# Determine the variance, standard deviation, and interquartile range for product rating
variance_rating = round(df['stars'].var(), 2)
std_dev_rating = round(df['stars'].std(), 2)
interquartile_range_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print("Variance Rating:", variance_rating)
print("Standard Deviation Rating:", std_dev_rating)
print("Interquartile Range Rating:", interquartile_range_rating)
```

```python
#Are the ratings consistent, or is there a wide variation in customer feedback?
# Ans: there is moderate variability in the ratings. It implies that ratings are not centered around the mean indicating some degree of variation in customer feedback.
```

```python
# Calculate the skewness and kurtosis for the rating column
skewness_rating = round(df['stars'].skew(), 2)
kurtosis_rating = round(df['stars'].kurtosis(), 2)

print("Skewness of Rating:", skewness_rating)
print("Kurtosis of Rating:", kurtosis_rating)
```

```python
# Are the ratings normally distributed, or do they lean towards higher or lower values?
# Ans: skewness of 0.19 suggests a slight right skew, indicating that the distribution of ratings are closer to higher values.
#  In this case, a kurtosis of -1.9 suggests a distribution is flatter than a normal distribution. 
#This indicates that the distribution has less extreme values compared to a normal distribution.
```

```python
# Plot a histogram to visualize the distribution of product ratings
plt.figure(figsize=(10, 6))
plt.hist(df['stars'], bins='auto', color='red', edgecolor='white')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
